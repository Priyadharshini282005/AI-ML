```python
import pandas as pd
```

```python
import numpy as np

```

```python
import matplotlib.pyplot as plt
```

```python
import random
```

```python
employee_ids = [f"EMP{str(i).zfill(3)}" for i in range(1, 31)]

```

```python
working_hours = [round(random.uniform(4, 10), 2) for _ in range(30)]
```

```python
df = pd.DataFrame({
    'EmployeeID': employee_ids,
    'WorkingHours': working_hours
})
```

```python
print("First 5 Records:\n", df.head())
print("\nSummary Statistics:\n", df.describe())

```

```python
departments = ['HR', 'Tech', 'Sales']
df['Department'] = [random.choice(departments) for _ in range(30)]
```

```python
grouped = df.groupby('Department')['WorkingHours'].mean().reset_index()
print("\nAverage Working Hours by Department:\n", grouped)
```

```python
df['Anomaly'] = df['WorkingHours'].apply(lambda x: 'Yes' if x < 5 or x > 9.5 else 'No')
print("\nDetected Anomalies:\n", df[df['Anomaly'] == 'Yes'])
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['EmployeeID'], df['WorkingHours'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Employee ID')
plt.ylabel('Working Hours')
plt.title('Working Hours of 30 Employees')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(6, 4))
plt.bar(grouped['Department'], grouped['WorkingHours'], color='lightgreen')
plt.xlabel('Department')
plt.ylabel('Avg Working Hours')
plt.title('Average Working Hours by Department')
plt.tight_layout()
plt.show()
```
